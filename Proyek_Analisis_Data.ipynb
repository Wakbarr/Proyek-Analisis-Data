## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana Kontribusi Masing-Masing Kategori Produk Terhadap Total Penjualan?

```python
# Menghitung jumlah order item per kategori produk (dalam bahasa Inggris)
sum_order_items_df = all_df.groupby("product_category_name_english").order_item_id.count().sort_values(ascending=False).reset_index()

# Menampilkan dataframe hasil pengelompokan
print(sum_order_items_df)
```

```python
# Membuat subplot untuk best performing dan worst performing product
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

# Warna khusus untuk kategori terbaik
colors = ["#72BF78", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

# Visualisasi kategori dengan jumlah penjualan tertinggi (Best Performing Product)
sns.barplot(x="order_item_id",
            y="product_category_name_english",
            data=sum_order_items_df.head(5),
            palette=colors,
            ax=ax[0])
ax[0].set_ylabel("Kategori Produk", fontsize=12)
ax[0].set_xlabel("Jumlah Penjualan (Order Item)", fontsize=12)
ax[0].set_title("Produk Terlaris", loc="center", fontsize=15)
ax[0].tick_params(axis='y', labelsize=12)

# Visualisasi kategori dengan jumlah penjualan terendah (Worst Performing Product)
sns.barplot(x="order_item_id",
            y="product_category_name_english",
            data=sum_order_items_df.sort_values(by="order_item_id", ascending=True).head(5),
            palette=colors,
            ax=ax[1])
ax[1].set_ylabel("Kategori Produk", fontsize=12)
ax[1].set_xlabel("Jumlah Penjualan (Order Item)", fontsize=12)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Produk dengan Penjualan Terendah", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)

plt.suptitle("Performa Penjualan Berdasarkan Kategori Produk", fontsize=20)
plt.tight_layout()
plt.show()
```

### Pertanyaan 2: Bagaimana Performa Pengiriman Berdasarkan Ketepatan Waktu (On Time vs. Late)?

```python
# Menghitung jumlah order per status pengiriman
delivery_status_df = order_df.groupby("delivery_status").order_id.count().reset_index()
delivery_status_df = delivery_status_df.sort_values(by="order_id", ascending=False)
print(delivery_status_df)
```

```python
# Membuat subplot untuk visualisasi performa pengiriman
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

# Subplot kiri: Bar chart jumlah order berdasarkan delivery_status
sns.barplot(x="order_id",
            y="delivery_status",
            data=delivery_status_df,
            palette=["#72BF78", "#D3D3D3"],
            ax=ax[0])
ax[0].set_ylabel("Status Pengiriman", fontsize=12)
ax[0].set_xlabel("Jumlah Order", fontsize=12)
ax[0].set_title("Performa Pengiriman (Jumlah Order)", fontsize=15, loc="center")
ax[0].tick_params(axis='y', labelsize=12)

# Subplot kanan: Pie chart distribusi persentase order berdasarkan delivery_status
colors = ["#72BF78", "#D3D3D3"]
ax[1].pie(delivery_status_df["order_id"],
          labels=delivery_status_df["delivery_status"],
          autopct='%1.1f%%',
          colors=colors,
          startangle=90)
ax[1].set_title("Distribusi Persentase Pengiriman", fontsize=15)

plt.suptitle("Evaluasi Performa Pengiriman Berdasarkan Ketepatan Waktu", fontsize=20)
plt.tight_layout()
plt.show()
```

**Insight:**
- Performa Pengiriman: Visualisasi bar chart di subplot kiri menunjukkan jumlah order yang dikirim tepat waktu (on time) dan terlambat (late). Jika jumlah order on time jauh lebih tinggi, ini mengindikasikan bahwa proses pengiriman berjalan dengan baik.
- Distribusi Persentase: Pie chart di subplot kanan memberikan gambaran persentase dari masing-masing status. Distribusi yang seimbang atau dominasi order on time akan mengonfirmasi efektivitas sistem logistik. Jika terdapat proporsi late yang cukup tinggi, perusahaan perlu melakukan evaluasi lebih lanjut untuk mengidentifikasi dan memperbaiki faktor penyebab keterlambatan pengiriman.

## Analisis Lanjutan (Opsional)

### RFM Analisis

- Recency: parameter yang digunakan untuk melihat kapan terakhir seorang pelanggan melakukan transaksi.
- Frequency: parameter ini digunakan untuk mengidentifikasi seberapa sering seorang pelanggan melakukan transaksi.
- Monetary: parameter terakhir ini digunakan untuk mengidentifikasi seberapa besar revenue yang berasal dari pelanggan tersebut.

```python
rfm_df = all_df.groupby(by="customer_unique_id", as_index=False).agg({
    "order_purchase_timestamp": "max",
    "order_id": "nunique",
    "payment_value": "sum"
})

rfm_df.columns = ["customer_unique_id", "max_order_timestamp", "frequency", "monetary"]

rfm_df["max_order_timestamp"] = pd.to_datetime(rfm_df["max_order_timestamp"]).dt.date

recent_date = all_df["order_purchase_timestamp"].dt.date.max()

rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

rfm_df.drop("max_order_timestamp", axis=1, inplace=True)

rfm_df.head()
```

```python
rfm_df.sort_values(by="frequency", ascending=False)
```

```python
rfm_df.sort_values(by="monetary", ascending=False)
```

```python
rfm_df.sort_values(by="recency", ascending=True)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))

colors = ["#72BF78", "#72BF78", "#72BF78", "#72BF78", "#72BF78"]

sns.barplot(y="recency", x="customer_unique_id", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=15)
plt.setp(ax[0].get_xticklabels(), rotation=90)

sns.barplot(y="frequency", x="customer_unique_id", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)
plt.setp(ax[1].get_xticklabels(), rotation=90)

sns.barplot(y="monetary", x="customer_unique_id", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)
plt.setp(ax[2].get_xticklabels(), rotation=90)

plt.suptitle("Best Customer Based on RFM Parameters (customer_unique_id)", fontsize=20)
plt.show()
```
